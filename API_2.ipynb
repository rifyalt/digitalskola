# Library

```
import pandas as pd, requests
```

```
url = 'http://api.open-notify.org/astros.json'
response = requests.get(url)
```

```
response.status_code
```

```
dict1 = response.json()
dict1
```

```
dict1['people'][1]['name']
```

# Data Processing

## Looping

```
dict1
```

```
message = dict1['message']
number = dict1['number']
```

```
temp = []
for obj in dict1['people']:
    craft = obj['craft']
    name = obj['name']
    row = [message, number, craft, name]
    temp.append(row)
df1 = pd.DataFrame(temp, columns=['message', 'number', 'craft', 'name'])
df1
```

## Cross Join

```
dict1
```

```
dict_base = {'message': [dict1['message']], 'number': [dict1['number']]}
base = pd.DataFrame(dict_base)
base
```

```
people = pd.DataFrame(dict1['people'])
people
```

```
# Add Key
base['key'] = 1
people['key'] = 1
df2 = pd.merge(base, people, on='key').drop('key', axis=1)
df2
```

# Alphavantage

```
url_alpha = 'https://www.alphavantage.co/query?'
queries = {
    'function': 'TIME_SERIES_DAILY_ADJUSTED',
    'symbol': 'GOOG',
    'apikey': 'XVV7I3GHUW42LTO4'
}
response = requests.get(url_alpha, params=queries)
response.status_code
```

```
stocks = response.json()
stocks
```

## Looping

```
symbol = stocks['Meta Data']['2. Symbol']
timezone = stocks['Meta Data']['5. Time Zone']

timeseries = stocks['Time Series (Daily)']
```

```
dates = list(timeseries.keys())
temp = []
for date in dates:
    open      = timeseries[date]['1. open']
    high      = timeseries[date]['2. high']
    low       = timeseries[date]['3. low']
    close     = timeseries[date]['4. close']
    adj_close = timeseries[date]['5. adjusted close']
    vol       = timeseries[date]['6. volume']
    div       = timeseries[date]['7. dividend amount']
    split     = timeseries[date]['8. split coefficient']
    row = [symbol, timezone, date, open, low, high, close, adj_close, vol, div, split]
    temp.append(row)
df3 = pd.DataFrame(temp, 
                   columns=['symbol', 'timezone', 'date', 'open', 'low', 'high',
                            'close', 'adj_close', 'vol', 'div', 'split'])
df3.head()
```

## Cross Join

```
symbol = stocks['Meta Data']['2. Symbol']
timezone = stocks['Meta Data']['5. Time Zone']
dict_base = {'symbol': [symbol], 'timezone': [timezone]}
base = pd.DataFrame(dict_base)
base
```

```
dict_time = stocks['Time Series (Daily)']
prices = pd.DataFrame(dict_time).T.reset_index()
prices.head()
```

```
base['key'] = 1
prices['key'] = 1
result = pd.merge(base, prices, on='key').drop('key', axis=1)
result.head()
```

```
result.columns = ['symbol', 'timezone', 'date', 'open', 'low', 'high', 'close', 'adj_close', 'vol', 'div', 'split']
result.head()
```

### CROSS JOIN
dipakai untuk dataset mirip dengan json, didalam objek ada objek lagi

```
# Define a dictionary containing user ID
dict1 = {'Name': ['Rebecca', 'Maryam', 'Anita'], 'UserID': [1, 2, 3]}

# Define
dict2 = {'ProductID': ['P1', 'P2', 'P3', 'P4']}

df = pd.DataFrame(dict1)
df1 = pd.DataFrame(dict2)
```

```
df['key'] = 1
df1['key'] = 1
```

```
df
```

```
cross = pd.merge(df, df1, on ='key').drop('key', 1)
# cross cukup pakai on berdasarkan apa, kasus disini mengunakan 'key' yang sejenis
```

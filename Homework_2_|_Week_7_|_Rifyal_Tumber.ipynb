```
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/gdrive') 
```

```
path_1 = 'gdrive/My Drive/Colab Notebooks//Digital Skola/order_details.csv' # urutan sesuai data disimpan pada grdive
path_2 = 'gdrive/My Drive/Colab Notebooks//Digital Skola/order_lists.csv' # urutan sesuai data disimpan pada grdive

details = pd.read_csv(path_1)
lists = pd.read_csv(path_2) 
```

```
details.head()
```

```
lists.head()
```

```
# Join Table
df = pd.merge(details, lists, how='right', left_on='order_id', right_on='order_id')
df
```

```
table = pd.pivot_table(data=df,index=['country'])
table
```

```
#  1. Berapakah Rata2 dan total penjualan dari setiap negara?

table1 = pd.pivot_table(df, values=['sales'], index=['country'],
                    aggfunc={'sales': [sum, np.mean]})
table1
```

```
# 2. Berapakah min dan max dari masing2 category produk?

table2 = pd.pivot_table(df, values=['Total_Cost'], index=['category'],
                    aggfunc={'Total_Cost': [min,max, np.sum]})
table2
```

```
# Berapakah total kontribusi profit dari setiap pembeli?

table3 = pd.pivot_table(df, values=['Total_profit'], index=['customer_name'],
                    aggfunc={'Total_profit': [np.sum]})
table3.head()
```

```
import requests
import pandas as pd
# from pandas.io.json import json_normalize
```

```
# https://covid19.mathdro.id/api/confirmed

url = "https://covid19.mathdro.id/api/confirmed"
response = requests.get(url)
```

```
response.status_code
```

```
response.json()
```

```
df = pd.read_json(url)
print(df)
```

```
covid19 = pd.DataFrame(df)
covid19.head(5)
```

```
covid19.filter(items = ['countryRegion', 'confirmed','recovered','deaths'] ).head(11)
```

```
covid19.filter(items = ['countryRegion', 'confirmed','recovered','deaths'] ).tail(11)
```

### Looping Saham Google

```
# berdasarkan symbol dan timezone
symbol = stocks['Meta Data']['2. Symbol']
timezone = stocks['Meta Data']['5. Time Zone']

timeseries = stocks['Time Series (Daily)']  
```

```
timeseries
```

```
timeseries['2020-09-28'] # untuk spesifik tanggal
```

```
dates = list(timeseries.keys())
temp =[]
for date in dates:
  open      = timeseries[date]['1. open']
  high      = timeseries[date]['2. high']
  low       = timeseries[date]['3. low']
  close     = timeseries[date]['4. close']
  adj_close = timeseries[date]['5. adjusted close']
  vol       = timeseries[date]['6. volume']
  div       = timeseries[date]['7. dividend amount']
  split     = timeseries[date]['8. split coefficient']
  row = [symbol, timezone, date, open, high, low, close, adj_close, vol, div, split]
  temp.append(row)
df3 = pd.DataFrame(temp, columns=['symbol', 'timezone', 'date', 'open', 'high', 'low', 'close', 'adj_close', 'vol', 'div', 'split'])
df3

```

```
# cara kedua dengan pandas

symbol = stocks['Meta Data']['2. Symbol']
timezone = stocks['Meta Data']['5. Time Zone']
dict_base = {'symbol':[symbol], 'timezone':[timezone]}
base = pd.DataFrame(dict_base)
base
```

```
dict_time = stocks['Time Series (Daily)']
prices = pd.DataFrame(dict_time).T # By Transpose
prices.head()
```

```
dict_time = stocks['Time Series (Daily)']
prices = pd.DataFrame(dict_time).T.reset_index() # By Transpose dan reset index agar tanggal jadi kolom
prices.head()
```

```
# cross join

base['key'] = 1
prices['key'] = 1

result = pd.merge(base, prices, on='key').drop('key', axis=1)
result.head()
```

```
# rapihin nama kolomnya
result.columns=['symbol', 'timezone', 'date', 'open', 'high', 'low', 'close', 'adj_close', 'vol', 'div', 'split']
result.head()
```
